# 2.IMPORT NECESSARY LIBRARIES

```python
import polars as pl
import pandas as pd
import selenium
import re
import sys
import time

# import selenium sub-modules and functions
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

#3.SETUP & FUNCTIONS

## i.Setup Chrome driver

```python
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.binary_location = '/usr/bin/chromium-browser'

chrome_options.add_argument("start-maximized")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

wd = webdriver.Chrome(options=chrome_options )
```

# 4.WEB-SCRAPE SAMSUNG PHONES LINKS

## i.Extract data from Alibaba website(Sections, Link, Topics & Content)

```python
wd = webdriver.Chrome(options=chrome_options )
url = "https://www.alibaba.com/trade/search?spm=a2700.product_home_newuser.home_new_user_first_screen_fy23_pc_search_bar.keydown__Enter&tab=all&SearchText=samsung+phone"  # Replace with the actual Konga URL for Infinix phones
wd.get(url)
print(wd.title)

  # Wait for the page to load
time.sleep(3)

links = wd.find_elements(By.XPATH,  "//h2/a")
l = []
for i in links:
  l.append(i.get_attribute("href"))
  print(i.get_attribute("href"))
print(l)
```

# 5.WEB-SCRAPE SAMSUNG PHONES

## i.Extract product_name from Alibaba Website

```python
product_names = []
for l1 in links:
  wd = webdriver.Chrome(options=chrome_options )
  wd.get(l1.get_attribute("href"))
  time.sleep(3)
  title = wd.find_elements(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div[1]/div/h1")
  for i in title:
    if len(i.text) == 0:
       product_names.append("None")
  else:
      print(i.text)
      product_names.append(i.text)
print(product_names)
```

## ii.Extract product_brand from Alibaba Website

```python
product_brand = []
for l2 in links:
  wd = webdriver.Chrome(options=chrome_options )
  wd.get(l2.get_attribute("href"))
  time.sleep(3)
  brand = wd.find_elements(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div[8]/div/div/div[1]/div[1]/div[2]/span")
  for i in brand:
    if len(i.text) == 0:
       product_brand.append("None")
  else:
      print(i.text)
      product_brand.append(i.text)
print(product_brand)
```

## iii.Extract product_price from Alibaba Website

```python
product_price = []
for l3 in links:
  wd = webdriver.Chrome(options=chrome_options )
  wd.get(l3.get_attribute("href"))
  time.sleep(3)
  price = wd.find_elements(By.XPATH, "/html/body/div[1]/div[1]/div[2]/div/div/div[5]/div/div/div[1]/div[2]/span")
  for i in price:
    if len(i.text) == 0:
       product_price.append("None")
  else:
      print(i.text)
      product_price.append(i.text)
      print(product_price)
```

# 6.CREATE A TABLE FOR THE EXTRACTED DATA

```python
import pandas as pd

# Assuming 'df' is the DataFrame created in the previous code
# Create a dictionary to store the data
data = {
    'product_name': product_names,
    'product_brand': product_brand,
    'Product_price': product_price,
    'product_link': l
}

# Create the Pandas DataFrame
df = pd.DataFrame(data)

# Convert the DataFrame to an HTML table
html_table = df.to_html()

#You can also display the table directly in a Jupyter Notebook or Colab environment:
from IPython.display import HTML
HTML(df.to_html())
```
